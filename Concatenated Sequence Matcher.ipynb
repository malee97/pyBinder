```python
import csv, os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
working_dir = os.path.join('/home','gridsan','mlee','ASMS_Analysis',
                           '20221129-329-077_X12K_2500_Members_Targeted_Seq_Test',
                          '329-077 X12K Targeted Test Updated 2022_12_01-08_44_08_AM')
prots = ['Aliq1','Aliq2']
rt_lim = 300

files_peaks = '329-077 X12K 2500 Member Concatenated.xlsx'
files_pyb = 'Consensus Features.csv'
output_file = '329-077 X12K PEAKS Matched to PyB.xlsx'
```

```python
peaks_seqs = pd.read_excel(os.path.join(working_dir,files_peaks),sheet_name='Concatenated')
feats_pyb = pd.read_csv(os.path.join(working_dir,files_pyb))
```

```python
peaks_seqs['Observed Mass'] = [np.round(mz*z - z*1.0073,3) for mz,z in zip(peaks_seqs['m/z'],peaks_seqs['z'])]
feats_pyb['Observed Mass'] = [np.round(mz*z - z*1.0073,3) for mz,z in zip(feats_pyb['m/z'],feats_pyb['charge'])]
```

```python
mass_peaks = peaks_seqs['Observed Mass'].to_list()
mass_pyb = feats_pyb['Observed Mass'].to_list()

match_idxs = []
for i,mass in enumerate(mass_pyb):
    if mass in mass_peaks:
        match_idxs.append(i)
        
match_idxs = np.unique(match_idxs)  # removes any duplicates, just in case
```

```python
found_mass = feats_pyb.iloc[match_idxs]
```

```python
found_mass
```

```python
seq_idxs = []
for j,mass in enumerate(mass_peaks):
    if mass in mass_pyb:
        seq_idxs.append(j)
        
seq_idxs = np.unique(seq_idxs)  # removes any duplicates, just in case
```

```python
found_seqs = (peaks_seqs.iloc[seq_idxs]
                .sort_values(by=['Observed Mass','ALC (%)'])
                .drop_duplicates(subset=['Observed Mass'],keep='last')
                .drop_duplicates(subset=['Peptide'],keep='last'))
```

```python
found_seqs
```

```python
with pd.ExcelWriter(os.path.join(working_dir,output_file)) as writer:
    found_seqs.to_excel(writer,sheet_name = 'From PEAKS')
    found_mass.to_excel(writer,sheet_name = 'From pyBinder')
```
